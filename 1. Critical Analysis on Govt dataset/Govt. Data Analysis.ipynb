```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

This code imports essential Python libraries: pandas (for data handling), NumPy (for numerical operations), matplotlib.pyplot (for plotting), and seaborn (for advanced visualizations). These libraries help in data analysis and visualization.

```python
df=pd.read_excel("Govt-dataset.xlsx")
df
```

```python
df.info()
```

This code loads an Excel file named "Govt-dataset.xlsx" into a pandas DataFrame called df and then displays its contents.

```python
df.columns = df.columns.str.strip().str.replace(r"[^\w\s]", "", regex=True).str.replace(" ", "_")
```

This code cleans the column names by removing extra spaces, special characters, and replacing spaces with underscores. It ensures column names are formatted consistently for easier access and analysis.

```python
df.drop(columns=["Unnamed_7", "Unnamed_11", "Unnamed_12"], errors="ignore", inplace=True)
```

This code removes the columns "Unnamed_7", "Unnamed_11", and "Unnamed_12" if they exist in the DataFrame. The errors="ignore" option prevents errors if the columns are not found.

```python
df
```

```python
numeric_cols = ["State__Code", "Number_of_towns", "Number_of_households", "Population", "Area\n_In_sq_km", "Population_per_sq_km"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

This code converts the specified columns to numeric data types, replacing any invalid values with NaN using errors='coerce'. This ensures that numerical operations can be performed on these columns.

```python
df = df.iloc[1:].reset_index(drop=True)
df
```

This code removes the first row from the DataFrame, resets the index, and drops the old index. It is useful if the first row contains metadata instead of actual data.

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df
```

This code fills missing values in numeric columns with their median values, ensuring the dataset remains complete for analysis while minimizing the impact of missing data.

```python
df=df.dropna()
df
```

This code removes all rows that contain any missing values, ensuring that the dataset has only complete records for analysis.

```python
df.isnull().sum()
```

This code checks for missing values in each column and returns the count of null values, helping to identify any remaining data gaps.

```python
plt.figure(figsize=(5,4))
sns.histplot(df['Population'], bins=30, kde=True, color='blue')
plt.title('Population Distribution')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

This code creates a histogram of the Population column with 30 bins and a density curve (kde=True), showing how population values are distributed. The blue color enhances visibility, and labels make the chart easy to understand.

```python
top_10_pop = df.nlargest(10, 'Population')
plt.figure(figsize=(5,4))
sns.barplot(x='Name', y='Population', data=top_10_pop, palette="viridis")
plt.title('Top 10 States/Districts by Population')
plt.xticks(rotation=45)
plt.show()
```

This code selects the top 10 states or districts with the highest population and creates a bar chart to visualize them. The x-axis represents the state or district names, the y-axis shows the population, and the colors are set using the viridis palette.

```python
urban_pop = df[df['Total\nRural\nUrban'] == 'Urban']['Population'].sum()
rural_pop = df[df['Total\nRural\nUrban'] == 'Rural']['Population'].sum()
labels = ['Urban', 'Rural']
sizes = [urban_pop, rural_pop]
colors = ['#ff9999','#66b3ff']
plt.figure(figsize=(5,4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Urban vs. Rural Population')
plt.axis('equal')  
plt.show()
```

This code calculates the total population for urban and rural areas, then creates a pie chart to compare them. The chart uses different colors for urban and rural populations, displays percentage labels, and ensures a circular shape with equal axis scaling.

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Area\n_In_sq_km', y='Population_per_sq_km', data=df, color='green')
plt.title('Population Density vs. Area')
plt.xlabel('Area (sq. km)')
plt.ylabel('Population per sq. km')
plt.show()
```

This code creates a scatter plot to show the relationship between area and population density. The x-axis represents the area in square kilometers, the y-axis shows the population per square kilometer, and green dots represent the data points.

```python
statewise_data = df.groupby('Name')[['Number_of_towns', 'Number_of_villages']].sum().nlargest(10, 'Number_of_towns')
statewise_data.plot(kind='bar', figsize=(5,4), color=['#ff9999', '#66b3ff'])
plt.title('Top 10 States by Number of Towns and Villages')
plt.xlabel('State/Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title="Category")
plt.show()
```

This code groups the data by state or region, sums the number of towns and villages, and selects the top 10 states with the most towns. It then creates a bar chart where towns and villages are represented in different colors, making it easy to compare their counts.

```python
plt.figure(figsize=(5,4))
sns.boxplot(df['Number_of_households'], color='lightcoral')
plt.title('Distribution of Number of Households')
plt.xlabel('Number of Households')
plt.show()
```

This code creates a box plot to show the distribution of the number of households. It helps identify the median, quartiles, and potential outliers in the dataset, with a light coral color for better visibility.

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='Number_of_households', y='Population', data=df, alpha=0.6, color='purple')
plt.title('Population vs. Number of Households')
plt.xlabel('Number of Households')
plt.ylabel('Population')
plt.show()
```

This code creates a scatter plot to show the relationship between the number of households and population. Each point represents a data entry, with transparency set to 0.6 for clarity, helping to identify trends or correlations.

```python
top_10_density = df.nlargest(10, 'Population_per_sq_km')
plt.figure(figsize=(5,4))
sns.barplot(y=top_10_density['Name'], x=top_10_density['Population_per_sq_km'], palette='coolwarm')
plt.title('Top 10 Most Densely Populated Regions')
plt.xlabel('Population per sq. km')
plt.ylabel('State/Region')
plt.show()
```

This code selects the top 10 most densely populated regions based on population per square kilometer and creates a horizontal bar chart. The x-axis represents population density, the y-axis lists the regions, and the coolwarm color palette enhances visualization.

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

This code generates a heatmap to show the correlation between numerical features in the dataset. The color intensity indicates the strength of relationships, with annotations displaying correlation values, helping to identify significant patterns or dependencies.

```python
plt.figure(figsize=(5,4))
plt.scatter(df['Area\n_In_sq_km'], df['Population'], s=df['Population_per_sq_km']/5, alpha=0.5, color='red')
plt.xlabel('Area (sq. km)')
plt.ylabel('Population')
plt.title('Bubble Chart of Area vs. Population')
plt.show()
```

The bubble size represents population density, making it easier to compare different states.

```python
g = sns.FacetGrid(df, col="Total\nRural\nUrban", height=5, aspect=1)
g.map_dataframe(sns.histplot, x="Population", bins=20, kde=True)
g.set_axis_labels("Population", "Frequency")
g.fig.suptitle('Rural vs. Urban Population Distribution', y=1.02)
plt.show()
```

This splits the dataset into rural and urban populations, showing their distribution separately.

```python
plt.figure(figsize=(5,4))
plt.hexbin(df['Population_per_sq_km'], df['Number_of_households'], gridsize=30, cmap='viridis', alpha=0.8)
plt.colorbar(label='Count')
plt.xlabel('Population Density (per sq. km)')
plt.ylabel('Number of Households')
plt.title('Hexbin Plot: Population Density vs. Households')
plt.show()
```

A hexbin plot helps visualize dense data points by grouping them into hexagonal bins.

```python
top_10 = df.nlargest(10, 'Population')
plt.figure(figsize=(5,4))
plt.stem(top_10['Name'], top_10['Population'], linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xticks(rotation=45)
plt.title('Top 10 States by Population (Lollipop Chart)')
plt.xlabel('State')
plt.ylabel('Population')
plt.show()
```

This **Lollipop Chart** highlights the top 10 most populated states using vertical lines and markers. It provides a cleaner alternative to bar charts, making it easier to compare population values.

```python
top_5_density = df.nlargest(5, 'Population_per_sq_km')
labels = top_5_density['Name']
values = top_5_density['Population_per_sq_km']

plt.figure(figsize=(5,4))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops={"edgecolor": "black"}, pctdistance=0.85)
plt.title('Top 5 Most Densely Populated States (Radial Chart)')
plt.show()
```

This Radial Chart (Pie Chart) shows the top 5 most densely populated states as percentage slices. It helps visualize how population density is distributed among the most crowded regions.

```python
plt.figure(figsize=(5,4))
sns.regplot(x='Area\n_In_sq_km', y='Population', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Regression Plot of Area vs. Population')
plt.xlabel('Area (sq. km)')
plt.ylabel('Population')
plt.show()
```

This Regression Plot shows the relationship between a state's area and its population. The red trend line indicates the overall correlation, helping to identify whether larger areas generally have higher populations.

```python
df_sorted = df.sort_values(by='Population_per_sq_km', ascending=False).head(10)
theta = np.linspace(0, 2*np.pi, len(df_sorted))
radii = df_sorted['Population_per_sq_km']
plt.figure(figsize=(5,4))
ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=0.4, color=plt.cm.magma(np.linspace(0, 1, len(df_sorted))), alpha=0.7)
plt.title('Circular Heatmap of Population Density')
plt.show()
```

This Circular Heatmap (Polar Bar Chart) represents the top 10 most densely populated states in a circular format. The bar heights and colors indicate population density, making it visually appealing for comparisons.

```python
plt.figure(figsize=(5,4))
sns.kdeplot(df['Population_per_sq_km'], fill=True, color='red', alpha=0.6)
plt.title('Kernel Density Estimation of Population Density')
plt.xlabel('Population per sq. km')
plt.ylabel('Density')
plt.show()
```

This **Kernel Density Estimation (KDE) Plot** shows the smooth distribution of population density. It helps visualize where most regions fall in terms of density, highlighting peaks and spread.

```python
sns.pairplot(df[['Population', 'Number_of_households', 'Area\n_In_sq_km', 'Population_per_sq_km']], diag_kind='kde', palette='husl')
plt.show()
```

This **Pair Plot** visualizes relationships between multiple numerical features using scatter plots and KDE plots. It helps identify correlations, trends, and distributions across different variables.

```python
plt.figure(figsize=(5,4))
plt.scatter(df['Area\n_In_sq_km'], df['Population'], s=df['Population_per_sq_km']/2, alpha=0.5, c=df['Population_per_sq_km'], cmap='coolwarm')
plt.colorbar(label='Population Density')
plt.title('Bubble Chart: Area vs. Population')
plt.xlabel('Area (sq. km)')
plt.ylabel('Population')
plt.show()
```

This **Bubble Chart** compares area and population, with bubble size representing population density. It visually highlights regions with high density, making it easy to spot outliers.

```python
top_10_states = df.nlargest(10, 'Population')

plt.figure(figsize=(5,4))
for i in range(len(top_10_states)):
    plt.plot([top_10_states['Number_of_households'].iloc[i], top_10_states['Population'].iloc[i]], 
             [i, i], color='gray', linewidth=2)
    plt.scatter(top_10_states['Number_of_households'].iloc[i], i, color='blue', label='Households' if i == 0 else "", s=100)
    plt.scatter(top_10_states['Population'].iloc[i], i, color='red', label='Population' if i == 0 else "", s=100)

plt.yticks(range(len(top_10_states)), top_10_states['Name'])
plt.xlabel('Count')
plt.title('Dumbbell Chart: Population vs. Number of Households (Top 10 States)')
plt.legend()
plt.show()
```

This Dumbbell Chart compares the difference between Population and Number of Households for the top 10 most populated states, making it easy to see the gap between the two values.
